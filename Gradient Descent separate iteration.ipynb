```python
import pandas as pd 
import numpy as np 
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples = 4, n_features = 1, n_informative = 1, n_targets = 1, noise=80, random_state=13)
```

```python
plt.scatter(X, y)
plt.show()
```

### Lets apply the OLS to find the value of slope(m).

```python
# here we find the m value to make the slope constant as we discuss in theory. 
from sklearn.linear_model import LinearRegression

lr = LinearRegression() 
lr.fit(X, y) 
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, lr.predict(X), color = 'red')
```

```python
# here we find the value of the slope(m). 
m = 78.35063668
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0
b = 0
y_pred = ((78.35 * X) + b).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
# first iteration. 
m = 78.35063668 
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
print(loss_slope)

# let we set the value of learning rate 
learning_rate = 0.1 

step_size = learning_rate * loss_slope
print(step_size)

b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```
